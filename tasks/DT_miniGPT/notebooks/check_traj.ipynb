```python
import json
import pandas as pd

def parse_json_to_dataframe(json_data):
    # Load the JSON data
    data = json.loads(json_data)

    # Create an empty list to store the parsed data
    parsed_data = []

    # Iterate over each item in the JSON data
    for item in data:
        instr_id = item['instr_id']
        trajectory = item['trajectory']

        # Iterate over each trajectory item and add instr_id and index
        for index, traj_item in enumerate(trajectory):
            parsed_item = [instr_id, index] + traj_item
            parsed_data.append(parsed_item)

    # Create a DataFrame from the parsed data
    columns = ['instr_id', 'step', 'id', 'value1', 'value2', 'value3']
    df = pd.DataFrame(parsed_data, columns=columns)

    return df


df = parse_json_to_dataframe(json_data)
print(df)
```
